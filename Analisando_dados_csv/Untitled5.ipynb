```python
df['date']=pd.to_datetime(df['date'])
df['dia'] = df['date'].dt.day
print(df['dia'])
```

```python
df2=df.groupby(df['dia']).agg({'timeonsite_st':'mean'}).reset_index()
print(df2)
```

```python
x1=df2['dia']
y1=df2['timeonsite_st']
mean = df2['timeonsite_st'].mean()
fig = plt.figure(figsize =(10, 5)) 
plt.plot(x1,y1)
plt.axhline(mean,xmin=0,xmax=31,color="r")
plt.grid()
plt.show()
```
